```python
import pandas as pd
data={'Color':['Red','Blue','Green','White']}
df=pd.DataFrame(data)
df_encoded=pd.get_dummies(df)
print(df_encoded)
```

```python
dt={'Age':[23,43,56,45,87,45,34,68,34]}
df=pd.DataFrame(dt)
bins=[0,20,40,60,100]
labels=['0-20','21-40','41-60','61+']

df['Age_Group']=pd.cut(df['Age'], bins=bins, labels=labels,right=False)
print(df)
```

```python
ds={'Full Address':[
    '123 Elm St, Springfield, 12345', '456 Oak Rd, Shelbyville, 67890']
}
df=pd.DataFrame(ds)
df[['Street','City','Zipcode']]=df['Full Address'].str.extract(
    r'([0-9]+\s[\w\s]+),\s([\w\s]+),\s(\d+)'
)
print(df)
```
