```python
import numpy as np
from sklearn.decomposition import FastICA
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
sample=200
time=np.linspace(0,8,sample)
signal1=np.sin(2*time)
signal2=np.sign(np.sin(3*time))
signal3=np.random.laplace(size=sample)
```

```python
S=np.c_[signal1,signal2,signal3]
S+=0.2*np.random.normal(size=S.shape)
A=np.array([[1, 1, 1], [0.5, 2, 1], [1.5, 1, 2]])
X=S.dot(A.T)
```

```python
ica=FastICA(n_components=3)
S_=ica.fit_transform(X)
```

```python
plt.figure(figsize=(8,6))
plt.subplot(3,1,1)
plt.title('Original_Source')
plt.show()
plt.subplot(3, 1, 2)
plt.title('Observed Signals')
plt.plot(X)

plt.subplot(3, 1, 3)
plt.title('Estimated Sources (FastICA)')
plt.plot(S_)

plt.tight_layout()
plt.show()
```
