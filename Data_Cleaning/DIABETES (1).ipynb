```python
fig,axs=plt.subplots(len(df.columns),1, figsize=(10,10),dpi=95)
for i,col in enumerate(df.columns):
  axs[i].boxplot(df[col],vert=False)
  axs[i].set_ylabel(col)
plt.tight_layout()
plt.show()
```
