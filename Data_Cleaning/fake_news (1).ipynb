```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('fake_news_dataset.csv')
df.head()
```

**Convert Attributes in uppercase and columns or
element into Lower case:**

```python
df.columns=[col.upper() for col in df.columns]
df.head()
```

```python
df[['TITLE', 'TEXT']] = df[['TITLE', 'TEXT']].apply(lambda x: x.str.lower())
df.head()
```

```python
df[['DATE','SOURCE','AUTHOR','CATEGORY','LABEL']]=df[['DATE','SOURCE','AUTHOR','CATEGORY','LABEL']].apply(lambda x: x.str.lower())
df.head()
```

Removing the emogies

```python
import re
emoji_pattern=re.compile(
    "["
    "\U0001F600-\U0001F64F"  # Emoticons
    "\U0001F300-\U0001F5FF"  # Symbols & Pictographs
    "\U0001F680-\U0001F6FF"  # Transport & Map Symbols
    "\U0001F700-\U0001F77F"  # Alchemical Symbols
    "]+", flags=re.UNICODE)

df['TITLE']=df['TITLE'].str.replace(emoji_pattern,'',regex=True)

df.head()
```

**Removel of hyperlink in columns python:**

```python
url_pattern=re.compile(r'https?://\S+|www\.\S+')
df['TITLE']=df['TITLE'].str.replace(url_pattern,' ',regex=True)
df.head()
```
