## Implement Euclidean Distance (k-NN Utility)

### Subtask:
Write a Python function `euclidean_distance(point1, point2)` that calculates the Euclidean distance between two data points (numpy arrays or lists).


**Reasoning**:
The subtask requires defining a Python function to calculate the Euclidean distance between two points. I will create a `code_block` that implements this function using `numpy` for efficient calculations.



```python
import numpy as np

def euclidean_distance(point1, point2):
    """
    Calculates the Euclidean distance between two data points.

    Args:
        point1 (np.array or list): The first data point.
        point2 (np.array or list): The second data point.

    Returns:
        float: The Euclidean distance between point1 and point2.
    """
    # Ensure points are numpy arrays for element-wise operations
    point1 = np.asarray(point1)
    point2 = np.asarray(point2)

    # Calculate the squared differences
    squared_diffs = (point1 - point2)**2

    # Sum the squared differences and take the square root
    distance = np.sqrt(np.sum(squared_diffs))

    return distance

print("Euclidean distance function defined successfully.")

# Example usage to verify the function
point_a = [1, 2, 3]
point_b = [4, 5, 6]
dist = euclidean_distance(point_a, point_b)
print(f"Euclidean distance between {point_a} and {point_b}: {dist}")

point_c = np.array([0.5, 1.0])
point_d = np.array([2.0, 3.0])
dist_np = euclidean_distance(point_c, point_d)
print(f"Euclidean distance between {point_c} and {point_d}: {dist_np}")
```

## Implement Manhattan Distance (k-NN Utility)

### Subtask:
Write a Python function `manhattan_distance(point1, point2)` that calculates the Manhattan distance between two data points.


**Reasoning**:
The subtask requires implementing a Python function to calculate the Manhattan distance between two data points. I will create a code block that defines this function using NumPy for efficient calculations, and then provide example usage to verify its correctness.



```python
import numpy as np

def manhattan_distance(point1, point2):
    """
    Calculates the Manhattan distance between two data points.

    Args:
        point1 (np.array or list): The first data point.
        point2 (np.array or list): The second data point.

    Returns:
        float: The Manhattan distance between point1 and point2.
    """
    # Ensure points are numpy arrays for element-wise operations
    point1 = np.asarray(point1)
    point2 = np.asarray(point2)

    # Calculate the absolute differences and sum them
    distance = np.sum(np.abs(point1 - point2))

    return distance

print("Manhattan distance function defined successfully.")

# Example usage to verify the function
point_a = [1, 2, 3]
point_b = [4, 5, 6]
dist = manhattan_distance(point_a, point_b)
print(f"Manhattan distance between {point_a} and {point_b}: {dist}")

point_c = np.array([0.5, 1.0])
point_d = np.array([2.0, 3.0])
dist_np = manhattan_distance(point_c, point_d)
print(f"Manhattan distance between {point_c} and {point_d}: {dist_np}")
```

## Implement k-NN Classifier Core Logic

### Subtask:
Write a Python function `predict_knn(train_X, train_y, test_point, k, distance_metric)` that takes training data, a single test point, the number of neighbors 'k', and a distance metric function. It should find the 'k' nearest neighbors to the `test_point` in `train_X` and return their corresponding labels from `train_y`. For now, use simple majority voting.


**Reasoning**:
The subtask requires implementing the core logic of the `predict_knn` function. I will create a code block that defines this function, calculates distances to the test point for all training points, identifies the k nearest neighbors, and then performs a majority vote to return the predicted label. I will also include an example usage to demonstrate its functionality.



```python
import numpy as np
from collections import Counter

def predict_knn(train_X, train_y, test_point, k, distance_metric):
    """
    Predicts the label for a single test point using the k-Nearest Neighbors algorithm.

    Args:
        train_X (pd.DataFrame or np.ndarray): Training features.
        train_y (pd.Series or np.ndarray): Training labels.
        test_point (np.ndarray or list): The single data point to classify.
        k (int): The number of nearest neighbors to consider.
        distance_metric (function): The function to use for calculating distance
                                    (e.g., euclidean_distance, manhattan_distance).

    Returns:
        int: The predicted label for the test point based on majority voting.
    """
    if not isinstance(train_X, np.ndarray):
        train_X = train_X.to_numpy()
    if not isinstance(train_y, np.ndarray):
        train_y = train_y.to_numpy()
    test_point = np.asarray(test_point)

    distances = []
    for i in range(len(train_X)):
        dist = distance_metric(train_X[i], test_point)
        distances.append((dist, train_y[i]))

    # Sort by distance
    distances.sort(key=lambda x: x[0])

    # Get k nearest neighbors' labels
    k_nearest_labels = [label for dist, label in distances[:k]]

    # Perform majority voting
    most_common = Counter(k_nearest_labels).most_common(1)
    return most_common[0][0]

print("k-NN prediction function 'predict_knn' defined successfully.")

# Example usage:
# Select a test point from the X_test set
sample_test_point = X_test.iloc[0].values

# Define k
k_neighbors = 5

# Predict using Euclidean distance
predicted_label_euclidean = predict_knn(X_train, y_train, sample_test_point, k_neighbors, euclidean_distance)
print(f"\nSample Test Point (first row of X_test):\n{sample_test_point[:5]}...")
print(f"Actual Label (for first row of X_test): {y_test.iloc[0]}")
print(f"Predicted Label (Euclidean, k={k_neighbors}): {predicted_label_euclidean}")

# Predict using Manhattan distance
predicted_label_manhattan = predict_knn(X_train, y_train, sample_test_point, k_neighbors, manhattan_distance)
print(f"Predicted Label (Manhattan, k={k_neighbors}): {predicted_label_manhattan}")
```

## Implement Weighted k-NN Prediction

### Subtask:
Modify the `predict_knn` function to incorporate weighted voting. Instead of simple majority, the vote of each neighbor is weighted by the inverse of its distance to the `test_point`.


**Reasoning**:
The subtask requires defining a new function `predict_knn_weighted` that implements weighted voting for k-NN. I will create a code block that defines this function, calculates inverse-distance weights for the k nearest neighbors, sums these weights by class, and returns the class with the highest total weight. I will also include example usage to demonstrate its functionality with both Euclidean and Manhattan distances.



```python
import numpy as np
from collections import Counter

def predict_knn_weighted(train_X, train_y, test_point, k, distance_metric):
    """
    Predicts the label for a single test point using the weighted k-Nearest Neighbors algorithm.

    Args:
        train_X (pd.DataFrame or np.ndarray): Training features.
        train_y (pd.Series or np.ndarray): Training labels.
        test_point (np.ndarray or list): The single data point to classify.
        k (int): The number of nearest neighbors to consider.
        distance_metric (function): The function to use for calculating distance
                                    (e.g., euclidean_distance, manhattan_distance).

    Returns:
        int: The predicted label for the test point based on weighted voting.
    """
    if not isinstance(train_X, np.ndarray):
        train_X = train_X.to_numpy()
    if not isinstance(train_y, np.ndarray):
        train_y = train_y.to_numpy()
    test_point = np.asarray(test_point)

    distances_labels = []
    for i in range(len(train_X)):
        dist = distance_metric(train_X[i], test_point)
        distances_labels.append((dist, train_y[i]))

    # Sort by distance
    distances_labels.sort(key=lambda x: x[0])

    # Get k nearest neighbors' distances and labels
    k_nearest = distances_labels[:k]

    # Calculate weighted votes
    weighted_votes = Counter()
    epsilon = 1e-6 # Small value to prevent division by zero

    for dist, label in k_nearest:
        if dist == 0:
            # If distance is zero, it's a perfect match, assign a very high weight
            # This neighbor essentially determines the prediction
            return label
        weight = 1 / (dist + epsilon) # Add epsilon to avoid division by zero for very small distances
        weighted_votes[label] += weight

    # Return the label with the highest total weight
    if not weighted_votes: # Handle case where k_nearest is empty or all distances are infinite
        return None # Or raise an error, depending on desired behavior
    return weighted_votes.most_common(1)[0][0]

print("k-NN weighted prediction function 'predict_knn_weighted' defined successfully.")

# Example usage:
# sample_test_point, y_test.iloc[0], k_neighbors, euclidean_distance, manhattan_distance are available from previous steps

print(f"\nSample Test Point (first row of X_test):\n{sample_test_point[:5]}...")
print(f"Actual Label (for first row of X_test): {y_test.iloc[0]}")

# Predict using Euclidean distance with weighted voting
predicted_label_euclidean_weighted = predict_knn_weighted(X_train, y_train, sample_test_point, k_neighbors, euclidean_distance)
print(f"Predicted Label (Weighted Euclidean, k={k_neighbors}): {predicted_label_euclidean_weighted}")

# Predict using Manhattan distance with weighted voting
predicted_label_manhattan_weighted = predict_knn_weighted(X_train, y_train, sample_test_point, k_neighbors, manhattan_distance)
print(f"Predicted Label (Weighted Manhattan, k={k_neighbors}): {predicted_label_manhattan_weighted}")
```

## Implement Gini Impurity (Decision Tree Utility)

### Subtask:
Write a Python function `calculate_gini_impurity(groups, classes)` that calculates the Gini impurity for a given split, where `groups` are the partitioned data and `classes` are the unique target labels.


**Reasoning**:
The subtask requires implementing a Python function to calculate the Gini impurity. I will create a code block that defines this function according to the instructions, including handling empty groups and calculating weighted impurity. I will then provide example usage to demonstrate its functionality.



```python
import numpy as np
from collections import Counter

def calculate_gini_impurity(groups, classes):
    """
    Calculates the Gini impurity for a given split.

    Args:
        groups (list of lists): List of partitioned data, where each inner list represents a group.
                                Each item in the inner list is a target label.
        classes (list): List of unique class labels in the target variable.

    Returns:
        float: The Gini impurity for the split.
    """
    gini = 0.0
    total_size = float(sum([len(group) for group in groups]))

    for group in groups:
        group_size = float(len(group))
        if group_size == 0:
            continue

        score = 0.0
        # Calculate score for each class in the group
        class_counts = Counter(group)
        for class_val in classes:
            # Count instances of this class in the current group
            count = class_counts.get(class_val, 0)
            p = count / group_size
            score += p * p

        # Weight the group's impurity by its proportion of the total
        gini += (1.0 - score) * (group_size / total_size)
    return gini

print("Gini impurity function 'calculate_gini_impurity' defined successfully.")

# Example Usage:
# Assume target labels are 3, 4, 5, 6, 7, 8
all_classes = sorted(y_train.unique())

# Example 1: A perfect split (or close to it)
# Group 1: all class 5
# Group 2: all class 6
example_groups_1 = [[5, 5, 5, 5], [6, 6, 6, 6]]
impurity_1 = calculate_gini_impurity(example_groups_1, all_classes)
print(f"\nExample 1 Gini Impurity (perfect split): {impurity_1:.4f}")

# Example 2: A mixed split
# Group 1: mixture of 5 and 6
# Group 2: mixture of 5 and 6
example_groups_2 = [[5, 5, 6, 5], [6, 6, 5, 6]]
impurity_2 = calculate_gini_impurity(example_groups_2, all_classes)
print(f"Example 2 Gini Impurity (mixed split): {impurity_2:.4f}")

# Example 3: One group is empty
example_groups_3 = [[5, 5, 5], []]
impurity_3 = calculate_gini_impurity(example_groups_3, all_classes)
print(f"Example 3 Gini Impurity (one empty group): {impurity_3:.4f}")

# Example 4: A single group (no split yet, max impurity if classes are balanced)
example_groups_4 = [[5, 5, 5, 6, 6, 6]]
impurity_4 = calculate_gini_impurity(example_groups_4, all_classes)
print(f"Example 4 Gini Impurity (single mixed group): {impurity_4:.4f}")
```

## Implement Best Split Function (Decision Tree Utility)

### Subtask:
Write a Python function `get_best_split(dataset, features, classes)` that iterates through all features and all possible split points within each feature to find the best split that minimizes Gini impurity. This function should return the feature index, split value, and the two resulting data groups.


**Reasoning**:
The subtask requires implementing a `get_best_split` function that iterates through features and split points, partitions data, calculates Gini impurity, and returns the best split. I will create a code block to define this function, using `numpy` for efficient array operations and leveraging the previously defined `calculate_gini_impurity` function.



```python
import numpy as np

def get_best_split(dataset, features, classes):
    """
    Finds the best split point (feature and value) in a dataset that minimizes Gini impurity.

    Args:
        dataset (np.ndarray): The dataset, where each row is a data point and the last column
                              is the target label. For example, features + target.
        features (list): A list of feature indices (column numbers in the dataset) to consider for splitting.
                         These should not include the target column index.
        classes (list): A list of unique class labels in the target variable.

    Returns:
        dict: A dictionary containing 'feature_idx', 'split_value', and 'groups' (the two partitioned groups
              of target labels that result from the best split).
              Returns None if no valid split can be found (e.g., dataset is empty).
    """
    if len(dataset) == 0: # Handle empty dataset case
        return None

    best_gini = float('inf')
    best_feature_idx = -1
    best_split_value = -1
    best_groups = None

    n_features = dataset.shape[1] - 1 # Exclude the target column
    target_col_idx = n_features

    # Iterate through each feature column provided
    for feature_idx in features:
        # Get unique values in the current feature column to use as potential split points
        possible_split_values = np.unique(dataset[:, feature_idx])

        for split_value in possible_split_values:
            group_left_labels = []  # Contains target labels for left group
            group_right_labels = [] # Contains target labels for right group

            # Partition the dataset into two groups based on the current feature and split value
            for row in dataset:
                if row[feature_idx] < split_value:
                    group_left_labels.append(row[target_col_idx])
                else:
                    group_right_labels.append(row[target_col_idx])

            # Calculate Gini impurity for the current split
            # calculate_gini_impurity expects groups as a list of lists of target labels
            gini = calculate_gini_impurity([group_left_labels, group_right_labels], classes)

            # Check if this split is better than the current best
            if gini < best_gini:
                best_gini = gini
                best_feature_idx = feature_idx
                best_split_value = split_value
                best_groups = [group_left_labels, group_right_labels]

    # If no valid split was found (e.g., all samples are identical or only one class in dataset),
    # best_gini might still be inf. In such cases, return None or an indication of no split.
    if best_gini == float('inf'):
        return None

    return {'feature_idx': best_feature_idx, 'split_value': best_split_value, 'groups': best_groups}

print("Function 'get_best_split' defined successfully.")

# Example Usage:
# Let's create a dummy dataset (features X, target y)
# Format: [feature1, feature2, target]
example_data = np.array([
    [1.0, 2.0, 5],
    [1.1, 2.1, 5],
    [1.5, 3.0, 6],
    [1.6, 3.1, 6],
    [2.0, 1.0, 5],
    [2.1, 1.1, 6],
    [0.9, 2.5, 5],
    [1.8, 2.8, 6]
])

# Assuming classes are available from previous steps (all_classes)
# If not, let's define them for this example
# all_classes = sorted(list(np.unique(example_data[:, -1]))) # Uncomment if all_classes is not globally available

# Features to consider for splitting (index 0 for feature1, index 1 for feature2)
feature_indices = [0, 1]

best_split_info = get_best_split(example_data, feature_indices, all_classes)

if best_split_info:
    print(f"\nBest Split Found:")
    print(f"  Feature Index: {best_split_info['feature_idx']}")
    print(f"  Split Value: {best_split_info['split_value']}")
    print(f"  Left Group (target labels): {best_split_info['groups'][0]}")
    print(f"  Right Group (target labels): {best_split_info['groups'][1]}")
else:
    print("\nCould not find a valid split.")
```

## Implement Simple Decision Tree (from scratch)

### Subtask:
Construct a recursive function `build_tree(node, max_depth, min_size, depth)` that builds a single decision tree. This function will use `get_best_split` at each node, recursively building left and right children. Include stopping conditions for `max_depth` and `min_size` (minimum samples required to split a node). Implement a prediction function for this tree.


**Reasoning**:
The subtask requires implementing the recursive `build_tree` function, a helper `to_terminal` function, and a `predict_tree` function. I will place all these components into a single code block, along with example usage to demonstrate building a tree and making a prediction.



```python
import numpy as np
from collections import Counter

# 1. Define a helper function to determine the prediction for a leaf node
def to_terminal(group):
    """
    Determines the predicted class for a leaf node based on the majority class in the group.
    Assumes group is not empty.
    """
    counts = Counter(group)
    return counts.most_common(1)[0][0]

# 2. Define the main recursive function to build the decision tree
def build_tree(dataset, features, classes, max_depth, min_size, depth):
    """
    Recursively builds a decision tree.

    Args:
        dataset (np.ndarray): The current subset of data (features + target).
                              The last column is assumed to be the target label.
        features (list): Indices of features (columns) to consider for splitting.
        classes (list): All unique class labels in the target variable.
        max_depth (int): Maximum depth of the tree.
        min_size (int): Minimum number of samples required to split a node.
        depth (int): Current depth of the node.

    Returns:
        dict or int: A dictionary representing an internal node, or an int representing a terminal node.
    """
    # If dataset is empty, it implies an invalid branch that shouldn't be reached with proper splitting.
    # However, for robustness, if an empty dataset is somehow passed, return None.
    if len(dataset) == 0:
        return None

    y_labels = dataset[:, -1].astype(int).tolist() # Extract target labels, convert to list of int

    # Stopping condition 1: Minimum size for a split
    if len(dataset) < min_size:
        return to_terminal(y_labels)

    # Stopping condition 2: Maximum depth reached
    if depth >= max_depth:
        return to_terminal(y_labels)

    # Stopping condition 3: All labels in the current dataset are the same (perfectly pure node)
    if len(np.unique(y_labels)) == 1:
        return to_terminal(y_labels)

    # Find the best split
    best_split = get_best_split(dataset, features, classes)

    # Stopping condition 4: No valid split found that improves Gini impurity
    if best_split is None:
        return to_terminal(y_labels)

    node = {
        'feature_idx': best_split['feature_idx'],
        'split_value': best_split['split_value']
    }

    left_group_data, right_group_data = best_split['groups']

    # Recursively build left and right children
    # If a group is empty after a split, that branch becomes a terminal node predicting
    # the majority class of the current (parent) node's labels.
    if len(left_group_data) == 0:
        node['left'] = to_terminal(y_labels) # If left group is empty, predict parent's majority
    else:
        node['left'] = build_tree(left_group_data, features, classes, max_depth, min_size, depth + 1)

    if len(right_group_data) == 0:
        node['right'] = to_terminal(y_labels) # If right group is empty, predict parent's majority
    else:
        node['right'] = build_tree(right_group_data, features, classes, max_depth, min_size, depth + 1)

    return node

# 3. Define a prediction function for this tree
def predict_tree(tree, row):
    """
    Predicts the label for a single data point using the trained decision tree.

    Args:
        tree (dict or int): The decision tree (root node), or a terminal node value.
        row (np.ndarray): The data point (features only) to predict.

    Returns:
        int: The predicted class label.
    """
    # If the current node is a terminal node (an integer label), return its value
    if not isinstance(tree, dict):
        return tree

    # Traverse based on the split condition
    if row[tree['feature_idx']] < tree['split_value']:
        return predict_tree(tree['left'], row)
    else:
        return predict_tree(tree['right'], row)

print("Decision tree building and prediction functions defined successfully.")

# Example Usage:
# Combine X_train_np and y_train_np into a single dataset
# Ensure y_train_np is 2D for stacking, or convert if needed
if y_train_np.ndim == 1:
    y_train_np_reshaped = y_train_np.reshape(-1, 1)
else:
    y_train_np_reshaped = y_train_np

# The dataset for build_tree needs to have features and the target in the last column.
# X_train_np contains features, y_train_np_reshaped contains target.
train_dataset_for_tree = np.hstack((X_train_np, y_train_np_reshaped))

# Define parameters for tree building
max_depth = 5
min_size = 10
feature_indices = list(range(X_train_np.shape[1])) # All feature columns

print(f"\nBuilding Decision Tree with max_depth={max_depth}, min_size={min_size}...")
my_decision_tree = build_tree(train_dataset_for_tree, feature_indices, all_classes, max_depth, min_size, 0)
print("Decision Tree built.")

# Make a prediction for a sample_test_point
sample_test_point_features = sample_test_point # sample_test_point is already a numpy array of features.

# Ensure `my_decision_tree` is not None before attempting prediction
if my_decision_tree is not None:
    predicted_label = predict_tree(my_decision_tree, sample_test_point_features)
    print(f"\nSample Test Point (first row of X_test features):\n{sample_test_point_features[:5]}...")
    print(f"Actual Label (for first row of X_test): {y_test.iloc[0]}")
    print(f"Predicted Label (Decision Tree): {predicted_label}")
else:
    print("\nDecision Tree could not be built (e.g., due to empty training data).")
```

## Implement Gaussian Naive Bayes (from scratch)

### Subtask:

I will put all these functions into a single code block and provide example usage to verify their correctness.



```python
import numpy as np
from collections import Counter

# 1. Helper function to calculate mean and standard deviation for each feature, for each class
def calculate_mean_std(X, y, classes):
    """
    Calculates the mean and standard deviation for each feature, for each class.

    Args:
        X (np.ndarray): Training features.
        y (np.ndarray): Training labels.
        classes (list): List of unique class labels.

    Returns:
        dict: A dictionary where keys are class labels and values are lists of (mean, std) tuples
              for each feature in that class.
    """
    statistics = {}
    for class_val in classes:
        # Get indices where the target label matches the current class
        X_class = X[y == class_val]
        class_stats = []
        for feature_idx in range(X.shape[1]):
            feature_data = X_class[:, feature_idx]
            mean = np.mean(feature_data)
            std = np.std(feature_data)
            class_stats.append((mean, std))
        statistics[class_val] = class_stats
    return statistics

# 2. Helper function to calculate the Gaussian Probability Density Function (PDF)
def gaussian_pdf(x, mean, std):
    """
    Calculates the Gaussian probability density function for a given value x.

    Args:
        x (float): The value for which to calculate the PDF.
        mean (float): The mean of the Gaussian distribution.
        std (float): The standard deviation of the Gaussian distribution.

    Returns:
        float: The probability density at value x.
    """
    # Handle cases where std is zero to prevent division by zero or infinite probabilities
    if std == 0:
        return 1.0 if x == mean else 0.0 # If std is 0, probability is 1 at mean, 0 elsewhere

    exponent = np.exp(-((x - mean)**2 / (2 * std**2)))
    return (1 / (np.sqrt(2 * np.pi) * std)) * exponent

# 3. Main function for Gaussian Naive Bayes prediction
def predict_gnb(train_X, train_y, test_point):
    """
    Predicts the label for a single test point using Gaussian Naive Bayes.

    Args:
        train_X (np.ndarray): Training features.
        train_y (np.ndarray): Training labels.
        test_point (np.ndarray): The single data point to classify.

    Returns:
        int: The predicted class label.
    """
    # Ensure inputs are numpy arrays
    if not isinstance(train_X, np.ndarray):
        train_X = train_X.to_numpy()
    if not isinstance(train_y, np.ndarray):
        train_y = train_y.to_numpy()
    if not isinstance(test_point, np.ndarray):
        test_point = np.asarray(test_point)

    unique_classes = sorted(np.unique(train_y))

    # Calculate statistics for each class and feature
    class_statistics = calculate_mean_std(train_X, train_y, unique_classes)

    # Calculate prior probabilities
    prior_probabilities = {}
    total_samples = len(train_y)
    for class_val in unique_classes:
        prior_probabilities[class_val] = np.sum(train_y == class_val) / total_samples

    # Calculate posterior probability for each class
    posterior_probs = {}
    for class_val, stats in class_statistics.items():
        prior = prior_probabilities[class_val]
        likelihood = 1.0
        for feature_idx in range(len(stats)):
            mean, std = stats[feature_idx]
            x = test_point[feature_idx]
            likelihood *= gaussian_pdf(x, mean, std)
        posterior_probs[class_val] = prior * likelihood

    # Return the class with the highest posterior probability
    best_class = None
    max_posterior = -1
    for class_val, prob in posterior_probs.items():
        if prob > max_posterior:
            max_posterior = prob
            best_class = class_val
    return best_class

print("Gaussian Naive Bayes functions defined successfully.")

# Example Usage:
# X_train_np, y_train_np, X_test, y_test, all_classes are available from previous steps

# Select a test point from the X_test set
sample_test_point_gnb = X_test.iloc[0].values

print(f"\nSample Test Point (first row of X_test):\n{sample_test_point_gnb[:5]}...")
print(f"Actual Label (for first row of X_test): {y_test.iloc[0]}")

# Predict using Gaussian Naive Bayes
predicted_label_gnb = predict_gnb(X_train_np, y_train_np, sample_test_point_gnb)
print(f"Predicted Label (Gaussian Naive Bayes): {predicted_label_gnb}")

# Evaluate GNB accuracy on a small subset of the validation set for demonstration
# (Full evaluation will be done in the next step to avoid long execution times in example)
num_val_samples = 10 # Predict for first 10 validation samples
y_pred_val_gnb = []
for i in range(num_val_samples):
    test_point_val = X_val_np[i]
    pred = predict_gnb(X_train_np, y_train_np, test_point_val)
    y_pred_val_gnb.append(pred)

from sklearn.metrics import accuracy_score
actual_val_labels = y_val_np[:num_val_samples]

# Handle potential None predictions if gaussian_pdf returns 0 everywhere for a class
# (e.g., if a class has zero std and the test point is not exactly at the mean)
# For now, let's assume predict_gnb always returns a class.
if None in y_pred_val_gnb:
    print("Warning: Some predictions were None. GNB may need further refinement for edge cases.")
    # Filter out None predictions for accuracy calculation, or assign a default class
    # For this example, if None is present, we'll indicate failure rather than calculate partial accuracy
    # A more robust GNB might have a smoothing parameter or different handling for 0 std.
    accuracy_gnb_subset = 'N/A due to None predictions'
else:
    accuracy_gnb_subset = accuracy_score(actual_val_labels, y_pred_val_gnb)

print(f"GNB Accuracy on first {num_val_samples} validation samples: {accuracy_gnb_subset}")
```
