```python
import pandas as pd
import numpy as np
```

```python
data = {
    'UserID': [101, 102, 103, 104, 105, 106, 107, 108],
    'Name': ['Alice Smith', 'Bob Johnson', 'Charlie Brown', 'David Lee', 'Eve Davis', 'Frank White', 'Grace Hall', 'Henry King'],
    'Score': [88.5, 95.0, np.nan, 72.3, 91.1, 88.5, 78.0, 95.0],
    'City': ['New York', 'London', 'Paris', 'New York', 'London', 'Tokyo', 'Paris', 'London'],
    'IsActive': [True, True, False, True, False, True, True, True],
    'Date': pd.to_datetime(['2024-01-01', '2024-01-02', '2024-01-03', '2024-01-04', '2024-01-05', '2024-01-06', '2024-01-07', '2024-01-08'])
}
```

```python
df = pd.DataFrame(data)
print(df)
```

```python
print("Score is greater than 90",df[df['Score']>90].to_string)
```

```python
filter_ny_active = (df['City'] == 'New York') & (df['IsActive'] == True)
print("Newyork and Isactive", df[filter_ny_active].to_string())
```

```python
print("Paris orTokyo", df.query('City == "Paris" or City == "Tokyo"').to_string())
```

```python
df_dropped = df.drop(columns=['IsActive'])
print("drop IsActive", df_dropped.shape)
```

```python
mean_score = df['Score'].mean()
df_filled = df.copy()
df_filled['Score'] = df_filled['Score'].fillna(mean_score)
print("\n22. df['Score'].fillna(mean_score) (Filled Score):\n", df_filled.loc[2, 'Score'])
```

```python
df_dropna = df.dropna()
print("missing value content  dropping ", df_dropna.shape)
```

```python
df_sorted = df.sort_values(by='Score', ascending=False)
print("Sorting dataframe", df_sorted[['Name', 'Score']].head(5).to_string())
```

```python
df['City'] = df['City'].replace('New York', 'NYC')
print("Replacing Newyork to NYC", df.loc[0, 'City'])
```

```python
print("Find Median", df['Score'].median())
```

```python
city_mean = df.groupby('City')['Score'].mean()
print("Mean Score by City", city_mean.to_string())
```

```python
score_agg = df.agg({'Score': ['count', 'std']})
print("\Coundand std", score_agg.to_string())
```

```python
print("Variance:", df['Score'].var())
```

```python
print("Quantile :", df['Score'].quantile(0.75))
```

```python
pivot_table = df.pivot_table(index='City', values=['Score', 'Name'], aggfunc={'Score': 'mean', 'Name': 'count'})
print("\n43. df.pivot_table(index='City', ...) (Pivot Table):\n", pivot_table.to_string())
```
