```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

**Data for Line/Scatter/Bar plots**

```python
np.random.seed(42)
X = np.linspace(0, 10, 50)
Y_sin = np.sin(X)
Y_cos = np.cos(X)
```

**Data for Categorical/Bar plots**

```python
categories = ['Apples', 'Bananas', 'Oranges', 'Grapes']
sales = np.array([30, 45, 25, 50])
```

**Data for Histograms/Statistical plots**

```python
hist_data = np.random.normal(loc=10, scale=2, size=1000)
box_data = [np.random.normal(0, std, 100) for std in range(1, 4)]
```

**Data for pie chart**

```python
market_share = [15, 30, 40, 15]
labels_pie = ['A', 'B', 'C', 'D']
```

```python
plt.figure(figsize=(6, 4))
plt.plot(X, Y_sin)
plt.title("1. Simple Line Plot (sin(x))")
```

```python
plt.figure(figsize=(6, 4))
plt.scatter(X, Y_cos)
plt.title("2. Simple Scatter Plot (cos(x))")
```

```python
plt.figure(figsize=(6, 4))
plt.bar(categories, sales)
plt.title("3. Vertical Bar Chart")
```

```python
plt.figure(figsize=(6, 4))
plt.barh(categories, sales)
plt.title("5. Horizontal Bar Chart")
```

```python
plt.figure(figsize=(6, 4))
plt.step(X, Y_sin, where='mid')
plt.title("6. Step Plot")
```

```python
plt.figure(figsize=(6, 4))
plt.fill_between(X, Y_sin, color='skyblue', alpha=0.4)
plt.title("7. Area Plot")
```

```python
plt.figure(figsize=(6, 4))
plt.plot(X, Y_sin, linestyle='--', linewidth=3, color='purple')
plt.title("11. Dashed Purple Line")
```

```python
plt.figure(figsize=(6, 4))
plt.plot(X, Y_sin)
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.title("21. Plot with Title and Labels")
```

```python
plt.figure(figsize=(6, 4))
plt.plot(X, Y_sin, label='Sine Wave')
plt.plot(X, Y_cos, label='Cosine Wave')
plt.legend(loc='upper right')
plt.title("22. Adding a Legend")
```

```python
plt.figure(figsize=(6, 4))
bars = plt.bar(categories, sales)
bars[1].set_color('green')
plt.title("26. Custom Bar Colors")
```

```python
plt.figure(figsize=(6, 4))
plt.plot(X, Y_sin)
plt.title("39. Plot saved to PNG (Not displayed)")
```

```python
plt.figure(figsize=(12, 3)) # Wide and short figure
plt.plot(X, Y_sin)
plt.title("40. Custom Figure Size (12x3)")
```

```python
plt.figure(figsize=(6, 4))
mean = [10, 15, 12]
std_dev = [0.8, 1.2, 0.5]
plt.errorbar(['A', 'B', 'C'], mean, yerr=std_dev, fmt='o', capsize=5)
plt.title("45. Error Bar Plot")
```

```python
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
scatter_x = np.random.rand(50)
scatter_y = np.random.rand(50)
scatter_z = np.random.rand(50)
ax.scatter(scatter_x, scatter_y, scatter_z, c='red', marker='^')
ax.set_title("49. 3D Scatter Plot")
```
