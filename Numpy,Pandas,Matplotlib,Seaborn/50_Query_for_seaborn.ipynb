```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
sns.set_theme(style="whitegrid")
```

```python
tips = sns.load_dataset("tips")
iris = sns.load_dataset("iris")
titanic = sns.load_dataset("titanic")
flights_data = sns.load_dataset("flights")
# Pivot for Heatmap
flights = flights_data.pivot(index="month", columns="year", values="passengers")
```

```python
plt.figure(figsize=(14, 10))
plt.suptitle("4 Essential Seaborn Plots (Simplified View)", fontsize=16, y=1.02)
```

```python
plt.subplot(2, 2, 1)
sns.scatterplot(x="total_bill", y="tip", data=tips)
plt.title("1. Scatter Plot (Bill vs Tip)")
```

```python
plt.subplot(2, 2, 2)
sns.boxplot(x="day", y="total_bill", data=tips)
plt.title("2. Box Plot (Bill by Day)")
```

```python
plt.subplot(2, 2, 3)
sns.histplot(data=tips, x="total_bill", kde=True)
plt.title("3. Histogram (Total Bill Distribution)")
```

```python
plt.subplot(2, 2, 4)
sns.regplot(x="total_bill", y="tip", data=tips)
plt.title("4. Regression Plot (Linear Trend)")
```

```python
sns.scatterplot(x="age", y="fare", style="sex", data=titanic)
```

```python
sns.countplot(x="deck", data=titanic)
```
