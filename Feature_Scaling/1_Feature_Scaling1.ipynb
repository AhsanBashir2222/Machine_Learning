```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

# Correctly re-run train_test_split to ensure X_train and X_test are feature DataFrames
X = df.drop('length_of_service', axis=1)
y = df['length_of_service']
X_train_df, X_test_df, y_train_series, y_test_series = train_test_split(X, y, test_size=0.3, random_state=0)

# Identify non-numeric columns in the feature set
# (Using X_train_df as a representative for feature columns)
non_numeric_cols = X_train_df.select_dtypes(include='object').columns.tolist()

# Drop non-numeric columns from the feature sets before scaling
X_train = X_train_df.drop(columns=non_numeric_cols, errors='ignore')
X_test = X_test_df.drop(columns=non_numeric_cols, errors='ignore')

# Initialize and apply StandardScaler to the cleaned numerical feature sets
scaler = StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
scaler.mean_
```
